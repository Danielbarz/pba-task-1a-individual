# Vector Space Model: Cosine Similarity
Notebook ini menunjukkan cara merepresentasikan teks sebagai vektor kata lalu menghitung kemiripan antar dokumen menggunakan **Cosine Similarity**.

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
```

## 1. Import Dataset Stanford

```python
corpus = [
    "Keluhan warga di kawasan Cilandak dan Pulomas sebelumnya mencuat akibat kebisingan aktivitas padel yang berlangsung hingga malam hari.", # CNN Indonesia
    "Warga Pulomas Jaktim Keluhkan Lapangan Padel Bising, Berujung Gugat ke PTUN.", # News Detik
    "Kelurahan Gandaria Selatan berhasil memediasi sengketa kebisingan antara pemilik lapangan padel dan warga di kawasan Haji Nawi Raya, Cilandak, Jakarta Selatan." # Metro TV News
]

vectorizer = CountVectorizer(vocabulary=['padel', 'bising'])
X = vectorizer.fit_transform(corpus)
```

## 2. Cosine Similarity

```python
# Membandingkan semua dokumen dengan Dokumen 0 (1 John)
cos_sim = cosine_similarity(X[0:1], X)

# Tampilkan Hasil
df = pd.DataFrame(X.toarray(), columns=['padel', 'bising'])
df['Cosine Sim to Doc 0'] = cos_sim.flatten()
print(df)
```

```python
import pandas as pd
from collections import Counter
tf = pd.DataFrame(
[Counter(doc.split()) for doc in documents],
).fillna(0)
tf
```

```python
import numpy as np
def length(v):
    return np.sqrt((v ** 2).sum())
def cos_dist(v, w):
    return 1 - (v * w).sum() / (length(v) * length(w))
cos_dist(tf.loc[0], tf.loc[1]), cos_dist(tf.loc[0], tf.loc[2])
```

```python
import pandas as pd
from collections import Counter

# Data dari slide
documents = [
    "padel bising sekali padel", # Doc 0
    "kebisingan padel",                # Doc 1
    "bising bising bising padel" # Doc 2
]

# Membuat DataFrame TF secara manual
tf = pd.DataFrame(
    [Counter(doc.split()) for doc in documents],
).fillna(0).astype(int)

# Menampilkan matriks
print("Matriks Term Frequency (TF):")
print(tf)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

full_cos_sim_matrix = cosine_similarity(tf)

# Display the full matrix as a DataFrame for better readability
cos_sim_df = pd.DataFrame(
    full_cos_sim_matrix,
    index=[f'Doc {i}' for i in range(len(documents))],
    columns=[f'Doc {i}' for i in range(len(documents))]
 )
print("Full Cosine Similarity Matrix:")
print(cos_sim_df)
```
